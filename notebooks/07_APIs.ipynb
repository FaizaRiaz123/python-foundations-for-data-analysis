```python
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
```

```python
bitcoin_data = cg.get_coin_market_chart_by_id(id='bitcoin',vs_currency='usd',days=30)
bitcoin_data
#response will be a JSON epxresses as a python dictionary of nested lists
```

```python
bitcoin_data['prices']
```

```python
#convert above data into dataframes
import pandas as pd
data = pd.DataFrame(bitcoin_data['prices'],columns=['timestamp','Price'])
data
```

```python
#to conver timestamp into understandable date
data['Date']=pd.to_datetime(data['timestamp'],unit='ms')
data
```
